```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

### Bar graph 

```python
plt.plot([1,2,3],[1,8,9])   # x-axis, y-axis
plt.show()
```

```python
# to add title 
plt.plot([1,2,3],[1,8,9])   # x-axis, y-axis
plt.title('Graph')
```

```python
# to add labels
plt.plot([1,2,3],[1,8,9])   # x-axis, y-axis
plt.title('Graph')
plt.xlabel('X items ')
plt.ylabel('Y items')
plt.show()
```

```python
# to add grid
plt.plot([1,2,3],[1,8,9])   # x-axis, y-axis
plt.title('Graph')
plt.xlabel('X items ')
plt.ylabel('Y items')
plt.grid()
```

```python
# some customization / legends
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y, 'b^--', label='2x')
plt.plot(x,y, 'ro-', label='3x')
```
